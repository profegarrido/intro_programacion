## Condicionales

El condicional *if* es una estructura de control fundamental en programación que permite ejecutar un bloque de código **solo si** una condición específica es verdadera (True). Actúa como un mecanismo de toma de decisiones: si la condición se cumple, se ejecuta el código; de lo contrario, se omite o se toma otra ruta (Else).

```text

SI imc >= 30
        rango = "Obesidad"
    SINO SI imc >= 25
        rango = "Sobrepeso"
    SINO SI imc >= 18.5
        rango = "Peso normal"
    SINO
        rango = "Peso bajo"
FIN SI

ESCRIBIR rango

```

```r
if(imc >= 30){
  rango = 'Obesidad'
} else if(imc >= 25){
  rango = 'Sobrepeso'
} else if(imc >= 18.5){
  rango = 'Saludable'
} else{
  rango = 'Bajo peso'
}

print(rango)
```

```r
moneda = 1

ifelse(moneda == 0, 'cara', 'sello')
```

## CICLOS

### FOR

El ciclo *for* es una estructura de control repetitiva (bucle) utilizada en programación para ejecutar un bloque de código un **número definido de veces**. Es ideal cuando se conoce de antemano el número de iteraciones, permitiendo inicializar una variable, establecer una condición de parada y actualizar el contador en una sola línea.

```text

PARA i = 1 HASTA 10 HACER
    MOSTRAR "Vuelta numero: ", i
FIN PARA
```

```r
for(i in 1:10){
    texto = paste0("Velta numero: ", i)
    print(texto)
}
```

### WHILE

El ciclo *while* (o bucle mientras) es una estructura de control repetitiva en programación que ejecuta un bloque de código continuamente mientras la **condición  sea verdadera** (true). Se utiliza comúnmente cuando el número de iteraciones no está definido de antemano y la condición se evalúa antes de cada iteración.

```text
i = 1
MIENTRAS i<10 HACER
    MOSTRAR "Vuelta numero: ", i
    i = i +1
FIN MIENTRAS
```

```r
i = 1
while(i <= 10){
    print("Vuelta numero ")
    i = i + 1
}
```

## Funciones

Una función en programación es un **bloque de código reutilizable y autónomo que realiza una tarea específica, como calcular un valor o procesar datos**. Permite organizar el programa en módulos más manejables, evitando repeticiones y facilitando el mantenimiento. Las funciones reciben datos de entrada (parámetros) y suelen devolver un resultado.

Las funciones en R son **bloques de código reutilizables y parametrizados que ejecutan tareas específicas, esenciales para estructurar, organizar y automatizar tareas**. Se crean mediante la palabra clave **function()**, aceptan argumentos de entrada, procesan datos y devuelven resultados, evitando la repetición de código y mejorando la legibilidad.
